```python
import pandas as pd
from faust.utilities import get_summary_df
umi9600 = pd.read_excel('../ScreenData/UMI9600/UMI9600.xlsx')
umi9600['Target Gene'] = umi9600['Construct IDs']
controls = [x for x in umi9600['Target Gene'].unique() if x.startswith('ONE_INTER') or x.startswith('NO_SITE')]
inputs = [x for x in umi9600.columns if x.startswith('Input') ]
outputs = [x for x in umi9600.columns if x.startswith('Tumor') or x.startswith('LN')]
umi9600_summary_df = get_summary_df(umi9600, controls, inputs, outputs, )
umi9600_summary_df['tissue'] = umi9600_summary_df['output_site'].apply(lambda x: x.split(' ')[0])
```

```python
from scipy.stats import skew, kurtosis
from scipy.stats import wilcoxon
```

```python
emp_means = []
emp_stds = []
emp_skews = []
emp_kurtosis = []
theo_means = []
theo_stds = []
theo_skews = []
theo_kurtosis = []
cols = []
for col in [x for x in umi9600.columns if 'Input' in x or 'LN' in x or 'Tumor' in x]:
    for construct in umi9600['Construct IDs'].unique():
        counts = umi9600[umi9600['Construct IDs']==construct][col].values

        import rpy2.robjects as robjects
        from rpy2.robjects.packages import importr
        from rpy2.robjects import pandas2ri
        import seaborn as sns
        import matplotlib.pyplot as plt
        import numpy as np
        #Must be activated
        pandas2ri.activate()
        importr('vcd')
        goodfit=robjects.r('goodfit')
        summary=robjects.r('summary')
        counts = robjects.IntVector(counts)
        #counts = np.random.negative_binomial(5,0.5, 100000)
        #counts = robjects.IntVector(counts)
        a=goodfit(counts,"nbinomial","ML")
        res = dict(zip(a.names, a))
        aa=summary(a)
        p = aa[0][2]
        df =  aa[0][1]
        cs =  aa[0][0]


        parameters = dict(zip(res['par'].names, res['par']))
        xrange = range(len(res['fitted']))

        nb_mean = (1-parameters['prob'][0])*parameters['size'][0]/(parameters['prob'][0])
        nb_var = (1-parameters['prob'][0])*parameters['size'][0]/((parameters['prob'][0])**2)
        nb_skew = (2-parameters['prob'][0])/np.sqrt((1-parameters['prob'][0])*parameters['size'][0])
        nb_ex_kurtosis = 6/parameters['size'][0] + ((parameters['prob'][0])**2)/((1-parameters['prob'][0])*parameters['size'][0])
        #print('fit', nb_mean, np.sqrt(nb_var))
        #print('empirical', np.mean(counts), np.std(counts))
        #plt.show()
        cols.append(col)
        emp_kurtosis.append(kurtosis(counts))
        emp_skews.append(skew(counts))
        emp_stds.append(np.std(counts))
        emp_means.append(np.mean(counts))
        theo_means.append(nb_mean)
        theo_stds.append(np.sqrt(nb_var))
        theo_skews.append(nb_skew)
        theo_kurtosis.append(nb_ex_kurtosis)
moments_df = pd.DataFrame(cols, columns=['site'])
moments_df['theo_mean'] = theo_means
moments_df['theo_std'] = theo_stds
moments_df['theo_skew'] = theo_skews
moments_df['theo_kurtosis'] = theo_kurtosis
moments_df['emp_mean'] = emp_means
moments_df['emp_std'] = emp_stds
moments_df['emp_skew'] = emp_skews
moments_df['emp_kurtosis'] = emp_kurtosis
fig, axes = plt.subplots(2,4,figsize=(8,4))
for i, moment in enumerate(['mean','std','skew','kurtosis']):
    axes[i%2,i//2].scatter(moments_df['theo_{}'.format(moment)], moments_df['emp_{}'.format(moment)],s=4)
    maxlim = 1.02*np.max([moments_df['theo_{}'.format(moment)].max(), moments_df['emp_{}'.format(moment)].max()])
    p = wilcoxon(moments_df['theo_{}'.format(moment)], moments_df['emp_{}'.format(moment)], correction=True).pvalue
    print(p)
    if i>0:
        axes[i%2,i//2].set_title('Wilcoxon p: {0:.2g}'.format(p))
    axes[i%2,i//2].set_xlim([0,1.02*maxlim])
    axes[i%2,i//2].set_ylim([0,1.02*maxlim])
    axes[i%2,i//2].plot([0,maxlim],[0,maxlim],alpha=0.2)
    axes[i%2,i//2].set_xlabel('modelled {}'.format(moment.replace('std','std. dev.')))
    axes[i%2,i//2].set_ylabel('true {}'.format(moment.replace('std','std. dev.')))
plt.tight_layout()

emp_means = []
emp_stds = []
emp_skews = []
emp_kurtosis = []
theo_means = []
theo_stds = []
theo_skews = []
theo_kurtosis = []
cols = []
for col in [x for x in umi9600.columns if 'Input' in x or 'LN' in x or 'Tumor' in x]:
    for construct in umi9600['Construct IDs'].unique():
        counts = umi9600[umi9600['Construct IDs']==construct][col].values
        import rpy2.robjects as robjects
        from rpy2.robjects.packages import importr
        from rpy2.robjects import pandas2ri
        import seaborn as sns
        import matplotlib.pyplot as plt
        import numpy as np
        #Must be activated
        pandas2ri.activate()
        importr('vcd')
        goodfit=robjects.r('goodfit')
        summary=robjects.r('summary')
        counts = robjects.IntVector(counts)
        #counts = np.random.negative_binomial(5,0.5, 100000)
        #counts = robjects.IntVector(counts)
        a=goodfit(counts,"nbinomial","MinChisq")
        res = dict(zip(a.names, a))
        aa=summary(a)
        p = aa[0][2]
        df =  aa[0][1]
        cs =  aa[0][0]


        parameters = dict(zip(res['par'].names, res['par']))
        xrange = range(len(res['fitted']))

        nb_mean = (1-parameters['prob'][0])*parameters['size'][0]/(parameters['prob'][0])
        nb_var = (1-parameters['prob'][0])*parameters['size'][0]/((parameters['prob'][0])**2)
        nb_skew = (2-parameters['prob'][0])/np.sqrt((1-parameters['prob'][0])*parameters['size'][0])
        nb_ex_kurtosis = 6/parameters['size'][0] + ((parameters['prob'][0])**2)/((1-parameters['prob'][0])*parameters['size'][0])
        #print('fit', nb_mean, np.sqrt(nb_var))
        #print('empirical', np.mean(counts), np.std(counts))
        #plt.show()
        cols.append(col)
        emp_kurtosis.append(kurtosis(counts))
        emp_skews.append(skew(counts))
        emp_stds.append(np.std(counts))
        emp_means.append(np.mean(counts))
        theo_means.append(nb_mean)
        theo_stds.append(np.sqrt(nb_var))
        theo_skews.append(nb_skew)
        theo_kurtosis.append(nb_ex_kurtosis)
moments_df = pd.DataFrame(cols, columns=['site'])
moments_df['theo_mean'] = theo_means
moments_df['theo_std'] = theo_stds
moments_df['theo_skew'] = theo_skews
moments_df['theo_kurtosis'] = theo_kurtosis
moments_df['emp_mean'] = emp_means
moments_df['emp_std'] = emp_stds
moments_df['emp_skew'] = emp_skews
moments_df['emp_kurtosis'] = emp_kurtosis
#fig, axes = plt.subplots(2,2,figsize=(8,8))
for j, moment in enumerate(['mean','std','skew','kurtosis']):
    i = j+4
    axes[i%2,i//2].scatter(moments_df['theo_{}'.format(moment)], moments_df['emp_{}'.format(moment)],s=4)
    maxlim = 1.02*np.max([moments_df['theo_{}'.format(moment)].max(), moments_df['emp_{}'.format(moment)].max()])
    p = wilcoxon(moments_df['theo_{}'.format(moment)], moments_df['emp_{}'.format(moment)], correction=True).pvalue
    axes[i%2,i//2].set_title('Wilcoxon p: {0:.2g}'.format(p))
    axes[i%2,i//2].set_xlim([0,1.02*maxlim])
    axes[i%2,i//2].set_ylim([0,1.02*maxlim])
    axes[i%2,i//2].plot([0,maxlim],[0,maxlim],alpha=0.2)
    axes[i%2,i//2].set_xlabel('modelled {}'.format(moment.replace('std','std. dev.')))
    axes[i%2,i//2].set_ylabel('true {}'.format(moment.replace('std','std. dev.')))

fig.suptitle('ML'+' '*70+'MinChisq', y=1.03,x=.55)
plt.tight_layout()
#fig.savefig("/home/smarkson/Dropbox (HMS)/FITS Paper/Figure 2/J-FitToNegativeBinomial/FitTwoNegativeBinomialAllGenes9Jan2023.pdf")
```
